# Act 3: The People Inside the System

In this act, the analysis moves from data cleaning to passenger behavior. The goal is to test whether patterns in the taxi system are only mechanical or whether they reflect human choices.


# Hypothesis

Longer taxi trips receive higher average tip amounts than shorter taxi trips.

This may happen because longer trips usually have higher fares, and passenger tips are often based on fare amount.
# Hypothesis and Statistical Test

The hypothesis tested in this act is:

**Longer taxi trips receive higher average tip amounts than shorter taxi trips.**

This is a behavioral hypothesis because tipping is influenced by passenger decisions, not only by the mechanical movement of taxis.

To test this hypothesis, the cleaned dataset is divided into two groups:

- Short trips: trips with distance less than or equal to the median trip distance
- Long trips: trips with distance greater than the median trip distance

The mean tip amount of both groups is compared.

A two-sample independent t-test is used because the goal is to compare the average tip amount between two independent groups of trips.

The hypotheses for the statistical test are:

Null hypothesis, H0:

There is no significant difference between the average tip amount of short trips and long trips.

Alternative hypothesis, H1:

There is a significant difference between the average tip amount of short trips and long trips.

A p-value less than 0.05 will be treated as evidence that the difference is statistically significant.


# Alternative Explanation

An alternative explanation for the same pattern is **payment type**.

Longer trips may appear to receive higher tips not only because passengers are more generous on longer trips, but because longer trips are more likely to be paid by card. Card payments record the tip directly in the dataset, while cash tips may not be fully recorded.

Therefore, if many short trips are paid in cash, their recorded tip amount may appear lower even when passengers actually tipped the driver. In this case, the observed difference in tip amounts would be partly caused by how tips are recorded, not only by passenger behavior.

Another possible explanation is airport trips. Airport trips are usually longer and more expensive, so they may naturally produce higher tip amounts. If many long trips are airport trips, then airport travel may be responsible for the higher tips rather than distance alone.


```python
act3_df = clean_df.copy()

act3_df["tip_percentage"] = (act3_df["tip_amount"] / act3_df["fare_amount"]) * 100

act3_df = act3_df[
    (act3_df["tip_percentage"] >= 0) &
    (act3_df["tip_percentage"] <= 100)
]

act3_df.shape
```

```python
median_distance = act3_df["trip_distance"].median()

short_trips = act3_df[act3_df["trip_distance"] <= median_distance]
long_trips = act3_df[act3_df["trip_distance"] > median_distance]

print("Median trip distance:", median_distance)
print("Short trips:", len(short_trips))
print("Long trips:", len(long_trips))
```

The median trip distance was **1.72 miles**.

Trips with distance less than or equal to 1.72 miles were classified as short trips.

Trips with distance greater than 1.72 miles were classified as long trips.

The dataset contained:

- Short trips: **4,580,929**
- Long trips: **4,565,619**


```python
print("Average tip amount for short trips:", short_trips["tip_amount"].mean())
print("Average tip amount for long trips:", long_trips["tip_amount"].mean())

print("Average tip percentage for short trips:", short_trips["tip_percentage"].mean())
print("Average tip percentage for long trips:", long_trips["tip_percentage"].mean())
```

The average tip amount for short trips was **$2.27**, while the average tip amount for long trips was **$5.24**.

This shows that long trips received a higher tip amount in dollars.

However, the average tip percentage for short trips was **24.20%**, while the average tip percentage for long trips was **18.84%**.

This means that although long trips received more money as tips, short trips received a higher tip relative to fare amount.


```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(
    short_trips["tip_amount"],
    long_trips["tip_amount"],
    equal_var=False
)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

# Statistical Test Result

The t-test result was:

T-statistic = **-1201.65**

P-value = **0.0**

Since the p-value is less than 0.05, the null hypothesis is rejected. This means there is a statistically significant difference between the average tip amount of short trips and long trips.

The average tip amount for short trips was **$2.27**, while the average tip amount for long trips was **$5.24**. Therefore, the result supports the hypothesis that longer trips receive higher average tip amounts.

However, the average tip percentage for short trips was **24.20%**, while the average tip percentage for long trips was **18.84%**. This shows that short trips receive a higher tip relative to the fare, even though long trips receive more tip money in total.


# Why This Explanation Is Stronger

The explanation that long trips receive higher tip amounts is supported by the data because the average tip amount for long trips was **$5.24**, while the average tip amount for short trips was only **$2.27**.

The t-test also showed a p-value of **0.0**, which is less than 0.05. This means the difference is statistically significant and is unlikely to be due to random variation.

However, this explanation is not complete by itself. The average tip percentage was higher for short trips (**24.20%**) than for long trips (**18.84%**). This means passengers on short trips may actually tip more generously relative to the fare.

Therefore, the stronger conclusion is:

**Long trips generate higher tip amounts mainly because they have higher fares, but short trips receive higher tips as a percentage of the fare.**

This explanation is stronger than simply saying "long trips get better tips" because it considers both dollar tip amount and tip percentage. It also acknowledges that payment type and airport trips may influence the pattern.


# Conditions That Would Change the Conclusion

This conclusion could change under several conditions.

First, the conclusion may change if only card payments are analyzed. Cash tips may not be fully recorded in the dataset, so including cash trips may underestimate the true tip amount for some trips.

Second, the conclusion may change if airport trips are removed. Airport trips are usually longer and more expensive, so they may increase the average tip amount for long trips. If the difference becomes smaller after removing airport trips, then airport travel may be driving the pattern.

Third, the conclusion may change if extreme outliers are removed more strictly. Very high fares or very high tips can increase the average tip amount for long trips.

Fourth, the conclusion may change if short and long trips are defined differently. In this analysis, the median distance of **1.72 miles** was used. If a different cutoff is used, the comparison between short and long trips may change.

Finally, the conclusion may change if tip percentage is treated as more important than tip amount. Long trips had higher tip amounts, but short trips had higher tip percentages. Therefore, the interpretation depends on whether the focus is driver earnings or passenger generosity.


# Competing Interpretation

A competing interpretation of the finding is that long trips do not receive higher tips because passengers behave differently. Instead, long trips may receive higher recorded tips because they are more likely to be expensive trips, airport trips, or card-payment trips.

In this interpretation, the pattern is not mainly about passenger generosity. It is about trip structure and payment recording. Longer trips naturally have higher fares, and if passengers tip by percentage, the dollar tip amount will also increase. In addition, card payments record tips directly, while cash tips may be missing or underreported in the dataset.

This competing interpretation is important because the results show that short trips actually had a higher average tip percentage: **24.20%** for short trips compared with **18.84%** for long trips. Therefore, long trips do not necessarily show more generous passenger behavior. They show higher tip amounts in dollars.

For this reason, my conclusion is not that passengers are more generous on long trips. A more accurate conclusion is that long trips produce higher recorded tip amounts because they have higher fares, while short trips show stronger tipping behavior when measured as a percentage of fare.


# Act 4: The System at Scale

In this act, individual trips are transformed into time-based summaries. Instead of studying one trip at a time, the dataset is aggregated by date, weekday, and hour to identify larger patterns in the taxi ecosystem.

The goal is to determine whether the system shows stable rhythms, repeated cycles, or random-looking noise.


```python
act4_df = clean_df.copy()

act4_df["pickup_date"] = act4_df["tpep_pickup_datetime"].dt.date
act4_df["pickup_hour"] = act4_df["tpep_pickup_datetime"].dt.hour
act4_df["pickup_day_name"] = act4_df["tpep_pickup_datetime"].dt.day_name()
act4_df["pickup_week"] = act4_df["tpep_pickup_datetime"].dt.isocalendar().week

act4_df[["tpep_pickup_datetime", "pickup_date", "pickup_hour", "pickup_day_name", "pickup_week"]].head()
```

```python
daily_summary = act4_df.groupby("pickup_date").agg(
    trip_count=("VendorID", "count"),
    average_fare=("fare_amount", "mean"),
    average_tip=("tip_amount", "mean"),
    average_distance=("trip_distance", "mean")
).reset_index()

daily_summary.head()
```

```python
hourly_summary = act4_df.groupby("pickup_hour").agg(
    trip_count=("VendorID", "count"),
    average_fare=("fare_amount", "mean"),
    average_tip=("tip_amount", "mean")
).reset_index()

hourly_summary
```

```python
plt.figure(figsize=(10, 5))
plt.plot(hourly_summary["pickup_hour"], hourly_summary["trip_count"], marker="o")
plt.title("Trip Volume by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Trips")
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()
```

```python
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

weekday_summary = act4_df.groupby("pickup_day_name").agg(
    trip_count=("VendorID", "count"),
    average_fare=("fare_amount", "mean"),
    average_tip=("tip_amount", "mean")
).reindex(weekday_order).reset_index()

weekday_summary
```

```python
plt.figure(figsize=(10, 5))
plt.bar(weekday_summary["pickup_day_name"], weekday_summary["trip_count"])
plt.title("Trip Volume by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Trips")
plt.xticks(rotation=45)
plt.show()
```

# Act 4 Pattern Interpretation

After aggregating the cleaned dataset by weekday and hour, the taxi system shows a clear rhythm rather than random movement.

The weekday pattern shows that trip volume is lowest on Sunday and Monday, then rises from Tuesday onward. The highest trip volume appears on Thursday, followed closely by Wednesday. This suggests that taxi demand is strongest during the middle of the work week and slightly weaker during weekends.

The hourly pattern is even stronger. Trip volume is very low during early morning hours, especially around 4 AM. After 5 AM, trip volume begins to rise steadily. Demand increases throughout the day and reaches its highest point around 6 PM. After that, trip volume begins to decline, although evening demand remains relatively high.

# Patterns That Persist

The clearest persistent pattern is the daily cycle. Taxi demand is low overnight, rises in the morning, stays high through the afternoon, and peaks in the evening. This pattern reflects the rhythm of city life, including work travel, errands, commuting, and evening activity.

The weekday pattern also persists. Midweek days, especially Wednesday and Thursday, show higher trip volume than Sunday and Monday.

# Patterns That Disappear

Individual trip-level randomness disappears after aggregation. At the raw trip level, the data may look messy and unpredictable, but when grouped by hour and weekday, clear patterns emerge.

Small variations between individual trips become less important once the system is viewed at scale.

# Patterns That May Be Artifacts

Some patterns may be caused by the way the data is grouped. For example, grouping all April, May, and June trips together by weekday may hide differences between individual months. A specific holiday, event, or weather condition could affect one date, but that effect may not be visible in the overall weekday graph.

Also, the hourly graph combines all days together. This means weekday rush-hour patterns and weekend nightlife patterns may be mixed into one average hourly pattern.

# Act 4 Conclusion

The NYC taxi ecosystem does not appear to move randomly. It shows a repeated time-based rhythm. Demand is lowest in the early morning, rises through the day, and peaks in the evening. It is also stronger during the middle of the week than on Sunday and Monday.

Therefore, the system appears cyclical and structured rather than pure noise. However, some patterns may depend on how the data is aggregated, so further analysis by month, weekday versus weekend, or holiday periods would be needed before calling the system fully stable.


# Act 5: Pressure on the System

In this act, the analysis moves from observation to simulation. The commissioner is considering a small policy shift, and the effect of this change must be tested using the initial seed value.

From Act 1, my seed value is:

r = 5

Therefore, I will simulate a **5% increase in fare amount**.

# Policy Change Chosen

The parameter selected for intervention is:

**fare_amount**

This parameter was chosen because fare amount directly affects passenger cost, driver revenue, total trip amount, and potentially tipping behavior. A small fare increase may appear simple, but at city scale it can create a large system-wide effect.


```python
act5_df = clean_df.copy()

r = 5
fare_increase_rate = r / 100

act5_df["new_fare_amount"] = act5_df["fare_amount"] * (1 + fare_increase_rate)

act5_df["fare_difference"] = act5_df["new_fare_amount"] - act5_df["fare_amount"]

act5_df[["fare_amount", "new_fare_amount", "fare_difference"]].head()
```

```python
old_total_fare = act5_df["fare_amount"].sum()
new_total_fare = act5_df["new_fare_amount"].sum()
fare_revenue_change = new_total_fare - old_total_fare

print("Old total fare revenue:", old_total_fare)
print("New total fare revenue after 5% increase:", new_total_fare)
print("Increase in fare revenue:", fare_revenue_change)
print("Percentage increase:", (fare_revenue_change / old_total_fare) * 100)
```

```python
old_average_fare = act5_df["fare_amount"].mean()
new_average_fare = act5_df["new_fare_amount"].mean()

print("Old average fare:", old_average_fare)
print("New average fare after 5% increase:", new_average_fare)
print("Average fare increase per trip:", new_average_fare - old_average_fare)
```

```python
act5_df["tip_percentage"] = act5_df["tip_amount"] / act5_df["fare_amount"]

act5_df["new_tip_amount"] = act5_df["new_fare_amount"] * act5_df["tip_percentage"]

old_total_tip = act5_df["tip_amount"].sum()
new_total_tip = act5_df["new_tip_amount"].sum()

print("Old total tips:", old_total_tip)
print("New total tips if tip percentage stays same:", new_total_tip)
print("Increase in tips:", new_total_tip - old_total_tip)
```

```python
act5_summary = pd.DataFrame({
    "Metric": [
        "Total fare revenue",
        "Average fare per trip",
        "Total tips"
    ],
    "Before policy change": [
        old_total_fare,
        old_average_fare,
        old_total_tip
    ],
    "After 5% fare increase": [
        new_total_fare,
        new_average_fare,
        new_total_tip
    ],
    "Change": [
        new_total_fare - old_total_fare,
        new_average_fare - old_average_fare,
        new_total_tip - old_total_tip
    ]
})

act5_summary
```

# Act 5 Interpretation

A 5% increase in fare amount was simulated because my seed value from Act 1 was r = 5.

The old total fare revenue was **$181,453,738.88**.

After the 5% fare increase, the new total fare revenue became **$190,526,425.82**.

The increase in fare revenue was **$9,072,686.94**, which is a **5% increase**.

The old average fare per trip was **$19.82**, while the new average fare per trip became **$20.81**.

The average fare increased by about **$0.99 per trip**.

This shows that the system absorbs the fare change mechanically because a 5% fare increase produces a direct 5% increase in fare revenue, assuming trip volume does not change.

However, the system may also amplify the change through tipping. The old total tips were **$34,589,904.92**. If passengers keep the same tip percentage after the fare increase, total tips increase to **$36,319,400.17**.

This creates an additional tip increase of **$1,729,495.25**.

Therefore, the effect of the fare increase is not limited to fare revenue. It also increases total recorded tips when tip behavior remains proportional to fare amount.

# Does the System Absorb, Amplify, or Respond Unexpectedly?

The system partly absorbs the change because the fare increase creates a predictable increase in fare revenue.

However, the system also amplifies the change through tips. Since tips are often connected to fare amount, a fare increase also increases tip amounts if passengers maintain the same tipping percentage.

There is no unexpected change in this simulation because trip volume was kept constant. But in the real world, passengers may respond differently. If taxis become more expensive, some passengers may reduce usage, switch to other transport options, or take shorter trips.

Therefore, the system appears stable under the simulation, but this stability depends on the assumption that passenger demand does not fall.

# Closing Frame: Are The Signs Real Or Just Noise?

This investigation began with a question: is the NYC taxi ecosystem functioning normally, or is it quietly drifting into inefficiency?

To answer this, every major decision in the analysis was made with a specific reason. The month range was not chosen randomly; it was derived from the roll number seed, which gave the investigation window of April to June. The dataset was taken from the official NYC TLC Trip Record Data source to avoid using pre-cleaned or curated data. Yellow Taxi Trip Records were selected because they represent the core taxi ecosystem being studied.

The data cleaning rules were also chosen deliberately. A real trip was defined as a trip with positive distance, positive fare, positive total amount, at least one passenger, positive duration, and a duration below 180 minutes. These rules removed records that could distort the analysis, such as impossible trips, invalid fares, and unrealistic durations.

The analysis also questioned its own assumptions. In Act 3, the initial finding showed that long trips had higher average tip amounts. A careless interpretation would be that passengers are more generous on long trips. However, the data showed that short trips had a higher tip percentage. This challenged the original interpretation and showed that long trips produce higher dollar tips mainly because fares are higher, not necessarily because passengers behave more generously.

The limits of the analysis are also important. The data can show recorded trips, fares, tips, distances, and time patterns, but it cannot fully explain passenger intentions, driver satisfaction, traffic conditions, weather, or unrecorded cash tips. The Act 5 simulation can show what happens if fares increase by 5%, but it cannot guarantee that passengers would continue taking the same number of trips.

Overall, the signs in the system appear real, but they do not show a simple collapse. The taxi ecosystem shows structure: demand is low overnight, rises through the day, peaks in the evening, and is stronger during midweek. This suggests a stable and cyclical rhythm rather than random noise.

At the same time, pressure exists. Evening demand, midweek concentration, fare sensitivity, and tipping differences show that the system is not perfectly stable. The better conclusion is that the NYC taxi ecosystem is not failing outright, but it is operating under visible pressure. The signs are real, but they must be interpreted carefully because the same data can produce misleading conclusions if cleaning choices, grouping methods, and assumptions are not questioned.
